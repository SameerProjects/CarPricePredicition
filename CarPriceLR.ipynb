```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cars_clean.csv', index_col=0)
pd.set_option('display.max_columns', None)
df
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
df.manufacturer.unique()
```

```python
def display_nunique(col):
    print(f'{col} uniques: {df[col].nunique()}')
    

display_nunique('engine')
display_nunique('transmission')
display_nunique('drivetrain')
display_nunique('fuel_type')
display_nunique('mpg')
display_nunique('exterior_color')
display_nunique('interior_color')
display_nunique('seller_name')
```

```python
sns.pairplot(data=df)
```

```python
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap="Blues")
```

```python
plt.figure(figsize=(20,8))
           
plt.subplot(1,2,1)
plt.title('Car Price Distribution Plot')
sns.distplot(df.price)

plt.subplot(1,2,2)
plt.title('Car Price Spread')
sns.boxplot(y=df.price)

plt.show()
```

Takeaways:

1. The distribution plot tells us that the data is right skewed indicating that the mean is greater than the median.    
2. This informs us that based on the dataset the mean overestimates the most common values in a postiviely skewed direction
    by an extraordinary amount so it looks like removing outliers is a possible next step

```python
bg = df.year.value_counts().sort_values(ascending=True)

bg.plot(kind='barh', figsize= (12,10))
plt.ylabel('Year')
plt.xlabel('Amount of Cars')
```

```python
bg = df.manufacturer.value_counts().sort_values(ascending=True)

bg.plot(kind='barh', figsize= (12,10))
plt.ylabel('Manufacturer')
plt.xlabel('Amount of Cars')
```

```python
pd.options.display.float_format = '{:20,.2f}'.format

df_manu = df.groupby('manufacturer')

manu_avg = df_manu['price'].mean().sort_values(ascending=True)

print(manu_avg)

manu_avg.plot(kind='barh', figsize=(12,8))
plt.ylabel('Manufacturer')
plt.xlabel('Amount of Cars')
plt.title('Average Cost of Car Based on Manufacturer');
```

Takeaways:

1. Very few values for cars from 2000 and prior. Might have to omit them from dataset due to lack of available data
2. Toyota and Ford seem to be the most popular companies, while Mitsubishi and Jaguar are the least
3. Porsche has the highest average cost by a significant margin. This definitely plays a role as to why our distribution is so
    positively skewed
